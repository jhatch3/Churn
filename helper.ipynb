```python
import math
import numpy as np

left = -5
right = 5

def sigmoid(x):
    1 / (1 + np.exp(-x))

def relu(x):
    return max(0,x)

def tanh(x):
    return math.tanh(x)
```

```python
import matplotlib.pyplot as plt

# Define the range
left = -5
right = 5
x = np.arange(left, right, 0.1)

# Activation functions
def relu(z):
    return np.maximum(0, z)

def sigmoid(z):
    return 1 / (1 + np.exp(-z))

def tanh(z):
    return np.tanh(z)

# Apply activation functions
relu_vals = relu(x)
sigmoid_vals = sigmoid(x)
tanh_vals = tanh(x)

# Optional: plot the functions
plt.plot(x, relu_vals, label="ReLU")
plt.plot(x, sigmoid_vals, label="Sigmoid")
plt.plot(x, tanh_vals, label="Tanh")
plt.legend()
plt.title("Activation Functions")
plt.grid(True)
plt.show()
```

```python
model = ["Decision Tree", "KNN", "SVM", "MLP"]
f1_scores = [0.92, 0.91, 0.91, 0.92]


def add_labels(x, y):
    for i in range(len(x)):
        plt.text(i, y[i] - .1, y[i], ha='center', fontweight="semibold")

add_labels(model, f1_scores)

plt.bar(model,f1_scores)
plt.title("F1 Scores For Each Tuned Model")
plt.xlabel("Model")
plt.ylabel("F1 Score")
plt.show()
```
